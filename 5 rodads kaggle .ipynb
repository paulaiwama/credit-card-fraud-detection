```python
# ============================================================
# COMPARAÇÃO CCNN, CFTNet e SICRNN - 5 RODADAS
# DATASET KAGGLE / ULB - CREDIT CARD FRAUD
# ============================================================

!pip install -q kagglehub imbalanced-learn tensorflow

# =========================
# Importações
# =========================
import os
import time
import random
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.stats import wilcoxon

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, average_precision_score,
    confusion_matrix, ConfusionMatrixDisplay,
    classification_report, roc_curve, auc
)
from sklearn.feature_selection import SelectKBest, f_classif

from imblearn.over_sampling import SMOTE

import kagglehub
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks
from tensorflow.keras.layers import (
    Input, Dense, Dropout, LayerNormalization,
    MultiHeadAttention, GlobalAveragePooling1D,
    Add, Reshape, GRU
)
from tensorflow.keras.models import Model, Sequential

# ============================================================
# Reprodutibilidade
# ============================================================

def fixar_seed(seed):
    np.random.seed(seed)
    random.seed(seed)
    tf.random.set_seed(seed)

fixar_seed(42)

# ============================================================
# 1) Baixar dataset
# ============================================================

path = kagglehub.dataset_download("mlg-ulb/creditcardfraud")
csv_path = os.path.join(path, "creditcard.csv")

print("Caminho do dataset:", path)
print("Arquivo CSV:", csv_path)

# ============================================================
# 2) Ler dataset
# ============================================================

df = pd.read_csv(csv_path)

print("\nShape do dataset:", df.shape)
display(df.head())

print("\nValores nulos:")
print(df.isnull().sum())

print("\nDistribuição da classe:")
print(df["Class"].value_counts())

print("\nDistribuição percentual:")
print((df["Class"].value_counts(normalize=True) * 100).round(4))

# ============================================================
# 3) Separar entrada e alvo
# ============================================================

X = df.drop(columns=["Class"]).copy()
y = df["Class"].astype(int).copy()

print("\nX:", X.shape)
print("y:", y.shape)

# ============================================================
# 4) Treino e teste
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

print("\nTreino:", X_train.shape, y_train.shape)
print("Teste :", X_test.shape, y_test.shape)

# ============================================================
# 5) Normalização
# ============================================================

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("\nX_train_scaled:", X_train_scaled.shape)
print("X_test_scaled :", X_test_scaled.shape)

# ============================================================
# 6) SMOTE
# ============================================================

smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train_scaled, y_train)

print("\nAntes do SMOTE:")
print(y_train.value_counts())

print("\nDepois do SMOTE:")
print(pd.Series(y_train_smote).value_counts())

print("\nX_train_smote:", X_train_smote.shape)
print("y_train_smote:", y_train_smote.shape)

# ============================================================
# 7) Preparação para CCNN
# ============================================================

def vetor_para_matriz_6x6(X_array):
    n_amostras = X_array.shape[0]
    X_img = np.zeros((n_amostras, 6, 6), dtype=np.float32)

    for i in range(n_amostras):
        vetor = X_array[i]
        matriz = np.zeros(36, dtype=np.float32)
        matriz[:30] = vetor
        X_img[i] = matriz.reshape(6, 6)

    return X_img

X_train_ccnn = vetor_para_matriz_6x6(X_train_smote)
X_test_ccnn = vetor_para_matriz_6x6(X_test_scaled)

X_train_ccnn = np.expand_dims(X_train_ccnn, axis=-1)
X_test_ccnn = np.expand_dims(X_test_ccnn, axis=-1)

print("\nX_train_ccnn:", X_train_ccnn.shape)
print("X_test_ccnn :", X_test_ccnn.shape)

# ============================================================
# 8) Modelo CCNN
# ============================================================

def construir_ccnn():
    model = models.Sequential([
        layers.Input(shape=(6, 6, 1)),
        layers.Conv2D(32, (2, 2), activation="relu", padding="same"),
        layers.Conv2D(64, (2, 2), activation="relu", padding="same"),
        layers.Dropout(0.2),
        layers.Conv2D(64, (2, 2), activation="relu", padding="same"),
        layers.Flatten(),
        layers.Dense(64, activation="relu"),
        layers.Dropout(0.3),
        layers.Dense(32, activation="relu"),
        layers.Dense(1, activation="sigmoid")
    ])

    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
        loss="binary_crossentropy",
        metrics=[
            "accuracy",
            tf.keras.metrics.Precision(name="precision"),
            tf.keras.metrics.Recall(name="recall"),
            tf.keras.metrics.AUC(name="auc")
        ]
    )

    return model

# ============================================================
# 9) Preparação para CFTNet
# ============================================================

def gerar_contrafactual_simples(X_fraude, n_amostras=None, ruido=0.02):
    if n_amostras is None:
        n_amostras = len(X_fraude)

    idx = np.random.choice(len(X_fraude), size=n_amostras, replace=True)
    base = X_fraude[idx].copy()
    perturbacao = np.random.normal(0, ruido, base.shape)

    X_cf = base + perturbacao
    X_cf = np.clip(X_cf, 0, 1)

    return X_cf

X_fraude = X_train_smote[y_train_smote == 1]
X_cf = gerar_contrafactual_simples(
    X_fraude,
    n_amostras=len(X_fraude),
    ruido=0.02
)

y_cf = np.ones(len(X_cf), dtype=int)

X_train_cft = np.vstack([X_train_smote, X_cf])
y_train_cft = np.hstack([y_train_smote, y_cf])

print("\nX_train_cft:", X_train_cft.shape)
print("y_train_cft:", y_train_cft.shape)

# ============================================================
# 10) Modelo CFTNet
# ============================================================

def build_cftnet(input_dim):
    entradas = Input(shape=(input_dim,))
    x = Reshape((input_dim, 1))(entradas)

    x_proj = Dense(32)(x)

    attn_output = MultiHeadAttention(
        num_heads=2,
        key_dim=16
    )(x_proj, x_proj)

    x = Add()([x_proj, attn_output])
    x = LayerNormalization()(x)

    ff = Dense(64, activation="relu")(x)
    ff = Dense(32)(ff)

    x = Add()([x, ff])
    x = LayerNormalization()(x)

    x = GlobalAveragePooling1D()(x)
    x = Dense(64, activation="relu")(x)
    x = Dropout(0.3)(x)

    saida = Dense(1, activation="sigmoid")(x)

    model = Model(inputs=entradas, outputs=saida)

    model.compile(
        optimizer="adam",
        loss="binary_crossentropy",
        metrics=[
            "accuracy",
            tf.keras.metrics.Precision(name="precision"),
            tf.keras.metrics.Recall(name="recall"),
            tf.keras.metrics.AUC(name="auc")
        ]
    )

    return model

# ============================================================
# 11) Preparação para SICRNN
# ============================================================

k_features = 20

selector = SelectKBest(score_func=f_classif, k=k_features)
X_train_sel = selector.fit_transform(X_train_smote, y_train_smote)
X_test_sel = selector.transform(X_test_scaled)

X_train_sicrnn = X_train_sel.reshape(
    (X_train_sel.shape[0], X_train_sel.shape[1], 1)
)

X_test_sicrnn = X_test_sel.reshape(
    (X_test_sel.shape[0], X_test_sel.shape[1], 1)
)

print("\nX_train_sicrnn:", X_train_sicrnn.shape)
print("X_test_sicrnn :", X_test_sicrnn.shape)

# ============================================================
# 12) Modelo SICRNN
# ============================================================

def build_sicrnn(input_shape):
    model = Sequential([
        layers.Input(shape=input_shape),
        GRU(64, return_sequences=False),
        Dropout(0.3),
        Dense(32, activation="relu"),
        Dropout(0.2),
        Dense(1, activation="sigmoid")
    ])

    model.compile(
        optimizer="adam",
        loss="binary_crossentropy",
        metrics=[
            "accuracy",
            tf.keras.metrics.Precision(name="precision"),
            tf.keras.metrics.Recall(name="recall"),
            tf.keras.metrics.AUC(name="auc")
        ]
    )

    return model

# ============================================================
# 13) Função de avaliação para as rodadas
# ============================================================

def rodar_avaliacao(nome_modelo, model, X_train, y_train, X_test, y_test, seed):
    early_stop = callbacks.EarlyStopping(
        monitor="val_loss",
        patience=5,
        restore_best_weights=True
    )

    reduce_lr = callbacks.ReduceLROnPlateau(
        monitor="val_loss",
        factor=0.5,
        patience=3,
        verbose=0
    )

    inicio = time.time()

    model.fit(
        X_train,
        y_train,
        validation_split=0.2,
        epochs=20,
        batch_size=256,
        callbacks=[early_stop, reduce_lr],
        verbose=0
    )

    tempo = time.time() - inicio

    y_proba = model.predict(X_test, verbose=0).ravel()
    y_pred = (y_proba >= 0.5).astype(int)

    return {
        "Seed": seed,
        "Dataset": "Kaggle",
        "Modelo": nome_modelo,
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred, zero_division=0),
        "Recall": recall_score(y_test, y_pred, zero_division=0),
        "F1-Score": f1_score(y_test, y_pred, zero_division=0),
        "ROC-AUC": roc_auc_score(y_test, y_proba),
        "PR-AUC": average_precision_score(y_test, y_proba),
        "Tempo (s)": tempo
    }

# ============================================================
# 14) Executar 5 rodadas por modelo
# ============================================================

seeds = [42, 7, 21, 123, 2025]
resultados_kaggle_5_rodadas = []

for seed in seeds:
    print(f"\n==================== KAGGLE | SEED {seed} ====================")

    fixar_seed(seed)

    # -------------------------
    # CCNN
    # -------------------------
    ccnn_model = construir_ccnn()

    resultado_ccnn = rodar_avaliacao(
        "CCNN",
        ccnn_model,
        X_train_ccnn,
        y_train_smote,
        X_test_ccnn,
        y_test,
        seed
    )

    resultados_kaggle_5_rodadas.append(resultado_ccnn)

    # -------------------------
    # CFTNet
    # -------------------------
    cftnet_model = build_cftnet(X_train_cft.shape[1])

    resultado_cftnet = rodar_avaliacao(
        "CFTNet",
        cftnet_model,
        X_train_cft,
        y_train_cft,
        X_test_scaled,
        y_test,
        seed
    )

    resultados_kaggle_5_rodadas.append(resultado_cftnet)

    # -------------------------
    # SICRNN
    # -------------------------
    sicrnn_model = build_sicrnn(
        (X_train_sicrnn.shape[1], X_train_sicrnn.shape[2])
    )

    resultado_sicrnn = rodar_avaliacao(
        "SICRNN",
        sicrnn_model,
        X_train_sicrnn,
        y_train_smote,
        X_test_sicrnn,
        y_test,
        seed
    )

    resultados_kaggle_5_rodadas.append(resultado_sicrnn)

# ============================================================
# 15) Resultados completos
# ============================================================

df_kaggle_5_rodadas = pd.DataFrame(resultados_kaggle_5_rodadas)

print("\n=== RESULTADOS COMPLETOS - KAGGLE - 5 RODADAS ===")
display(df_kaggle_5_rodadas.round(6))

df_kaggle_5_rodadas.to_csv(
    "kaggle_resultados_5_rodadas_completos.csv",
    index=False
)

# ============================================================
# 16) Média e desvio padrão
# ============================================================

resumo_kaggle = df_kaggle_5_rodadas.groupby("Modelo").agg({
    "Accuracy": ["mean", "std"],
    "Precision": ["mean", "std"],
    "Recall": ["mean", "std"],
    "F1-Score": ["mean", "std"],
    "ROC-AUC": ["mean", "std"],
    "PR-AUC": ["mean", "std"],
    "Tempo (s)": ["mean", "std"]
}).round(6)

print("\n=== RESUMO KAGGLE - MÉDIA E DESVIO PADRÃO ===")
display(resumo_kaggle)

resumo_kaggle.to_csv("kaggle_resumo_media_desvio.csv")

# ============================================================
# 17) Tabela formatada para artigo
# ============================================================

def formatar_media_desvio(df):
    linhas = []

    for modelo in df.index:
        linha = {"Modelo": modelo}

        for metrica in [
            "Accuracy",
            "Precision",
            "Recall",
            "F1-Score",
            "ROC-AUC",
            "PR-AUC",
            "Tempo (s)"
        ]:
            media = df.loc[modelo, (metrica, "mean")]
            desvio = df.loc[modelo, (metrica, "std")]
            linha[metrica] = f"{media:.4f} ± {desvio:.4f}"

        linhas.append(linha)

    return pd.DataFrame(linhas)

tabela_kaggle_publicacao = formatar_media_desvio(resumo_kaggle)

print("\n=== TABELA KAGGLE FORMATADA PARA PUBLICAÇÃO ===")
display(tabela_kaggle_publicacao)

tabela_kaggle_publicacao.to_csv(
    "kaggle_tabela_final_publicacao.csv",
    index=False
)

# ============================================================
# 18) Teste estatístico Wilcoxon - F1-Score
# ============================================================

print("\n=== TESTE ESTATÍSTICO WILCOXON - KAGGLE ===")

ccnn_f1 = df_kaggle_5_rodadas[
    df_kaggle_5_rodadas["Modelo"] == "CCNN"
]["F1-Score"].values

cftnet_f1 = df_kaggle_5_rodadas[
    df_kaggle_5_rodadas["Modelo"] == "CFTNet"
]["F1-Score"].values

sicrnn_f1 = df_kaggle_5_rodadas[
    df_kaggle_5_rodadas["Modelo"] == "SICRNN"
]["F1-Score"].values

stat_ccnn_sic, p_ccnn_sic = wilcoxon(ccnn_f1, sicrnn_f1)
stat_ccnn_cft, p_ccnn_cft = wilcoxon(ccnn_f1, cftnet_f1)
stat_sic_cft, p_sic_cft = wilcoxon(sicrnn_f1, cftnet_f1)

print(f"CCNN vs SICRNN   -> p-valor = {p_ccnn_sic:.6f}")
print(f"CCNN vs CFTNet   -> p-valor = {p_ccnn_cft:.6f}")
print(f"SICRNN vs CFTNet -> p-valor = {p_sic_cft:.6f}")

# ============================================================
# 19) Matriz de confusão do melhor modelo por F1-Score médio
# ============================================================

melhor_modelo = resumo_kaggle[("F1-Score", "mean")].idxmax()

print(f"\nMelhor modelo pelo F1-Score médio: {melhor_modelo}")

# Treinar novamente o melhor modelo com seed 42 apenas para gerar matriz
fixar_seed(42)

if melhor_modelo == "CCNN":
    modelo_final = construir_ccnn()
    Xtr_final = X_train_ccnn
    ytr_final = y_train_smote
    Xte_final = X_test_ccnn

elif melhor_modelo == "CFTNet":
    modelo_final = build_cftnet(X_train_cft.shape[1])
    Xtr_final = X_train_cft
    ytr_final = y_train_cft
    Xte_final = X_test_scaled

else:
    modelo_final = build_sicrnn(
        (X_train_sicrnn.shape[1], X_train_sicrnn.shape[2])
    )
    Xtr_final = X_train_sicrnn
    ytr_final = y_train_smote
    Xte_final = X_test_sicrnn

early_stop_final = callbacks.EarlyStopping(
    monitor="val_loss",
    patience=5,
    restore_best_weights=True
)

reduce_lr_final = callbacks.ReduceLROnPlateau(
    monitor="val_loss",
    factor=0.5,
    patience=3,
    verbose=0
)

modelo_final.fit(
    Xtr_final,
    ytr_final,
    validation_split=0.2,
    epochs=20,
    batch_size=256,
    callbacks=[early_stop_final, reduce_lr_final],
    verbose=0
)

y_proba_final = modelo_final.predict(Xte_final, verbose=0).ravel()
y_pred_final = (y_proba_final >= 0.5).astype(int)

cm = confusion_matrix(y_test, y_pred_final)

disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.title(f"Matriz de Confusão - Melhor Modelo: {melhor_modelo}")
plt.show()

print("\nRelatório de classificação do melhor modelo:")
print(classification_report(y_test, y_pred_final, digits=6))

# ============================================================
# 20) Arquivos gerados
# ============================================================

print("\nArquivos salvos:")
print("- kaggle_resultados_5_rodadas_completos.csv")
print("- kaggle_resumo_media_desvio.csv")
print("- kaggle_tabela_final_publicacao.csv")
```
