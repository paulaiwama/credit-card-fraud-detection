```python
# ============================================================
# COMPARAÇÃO: CCNN, CFTNet e SICRNN
# ============================================================

# 1) Instalação
!pip install -q kagglehub imbalanced-learn tensorflow

# 2) Importações
import os
import time
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, average_precision_score,
    confusion_matrix, ConfusionMatrixDisplay, classification_report,
    roc_curve, auc
)
from sklearn.feature_selection import SelectKBest, f_classif
from imblearn.over_sampling import SMOTE

import kagglehub

import tensorflow as tf
from tensorflow.keras import layers, models, callbacks
from tensorflow.keras.layers import (
    Input, Dense, Dropout, LayerNormalization,
    MultiHeadAttention, GlobalAveragePooling1D,
    Add, Reshape, GRU
)
from tensorflow.keras.models import Model, Sequential

# 3) Reprodutibilidade
np.random.seed(42)
tf.random.set_seed(42)

# ============================================================
# 4) Baixar dataset Sparkov
# ============================================================
path = kagglehub.dataset_download("kartik2112/fraud-detection")
print("Caminho do dataset:", path)

# listar arquivos disponíveis
for raiz, dirs, files in os.walk(path):
    for f in files:
        print(os.path.join(raiz, f))

# ============================================================
# 5) Ler arquivos e unir
# ============================================================
train_path = os.path.join(path, "fraudTrain.csv")
test_path  = os.path.join(path, "fraudTest.csv")

df_train = pd.read_csv(train_path)
df_test  = pd.read_csv(test_path)

df = pd.concat([df_train, df_test], ignore_index=True)

print("\nShape do dataset:", df.shape)
display(df.head())

# ============================================================
# 6) Verificar colunas e alvo
# ============================================================
print("\nColunas:")
print(df.columns.tolist())

print("\nValores nulos:")
print(df.isnull().sum())

print("\nDistribuição da classe:")
print(df["is_fraud"].value_counts())

print("\nDistribuição percentual:")
print((df["is_fraud"].value_counts(normalize=True) * 100).round(4))

# ============================================================
# 7) Remover colunas desnecessárias
# ============================================================
colunas_para_remover = [
    "Unnamed: 0", "trans_date_trans_time", "cc_num",
    "first", "last", "street", "city", "state", "zip",
    "dob", "trans_num", "unix_time", "merchant", "job"
]

df = df.drop(columns=colunas_para_remover, errors="ignore")

print("\nColunas restantes:")
print(df.columns.tolist())

# ============================================================
# 8) Transformar colunas categóricas em numéricas
# ============================================================
colunas_categoricas = df.select_dtypes(include=["object"]).columns.tolist()
print("\nColunas categóricas:", colunas_categoricas)

df = pd.get_dummies(df, columns=colunas_categoricas, drop_first=True)

print("Shape após get_dummies:", df.shape)

# ============================================================
# 9) Separar entrada e alvo
# ============================================================
TARGET_COL = "is_fraud"

X = df.drop(columns=[TARGET_COL]).copy()
y = df[TARGET_COL].astype(int).copy()

print("\nX:", X.shape)
print("y:", y.shape)

# ============================================================
# 10) Limpeza final
# ============================================================
X = X.replace([np.inf, -np.inf], np.nan)
X = X.dropna()
y = y.loc[X.index]

print("\nX após limpeza:", X.shape)
print("y após limpeza:", y.shape)

# ============================================================
# 11) Divisão treino e teste
# ============================================================
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

print("\nTreino:", X_train.shape, y_train.shape)
print("Teste :", X_test.shape, y_test.shape)

# ============================================================
# 12) Normalização
# ============================================================
scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("\nX_train_scaled:", X_train_scaled.shape)
print("X_test_scaled :", X_test_scaled.shape)

# ============================================================
# 13) SMOTE
# ============================================================
smote = SMOTE(random_state=42)

X_train_smote, y_train_smote = smote.fit_resample(X_train_scaled, y_train)

print("\nAntes do SMOTE:")
print(y_train.value_counts())

print("\nDepois do SMOTE:")
print(pd.Series(y_train_smote).value_counts())

print("\nX_train_smote:", X_train_smote.shape)
print("y_train_smote:", y_train_smote.shape)

# ============================================================
# 14) Função de métricas
# ============================================================
def calcular_metricas(y_true, y_pred, y_proba, tempo_treino, nome_modelo):
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, zero_division=0)
    rec = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)
    roc_auc = roc_auc_score(y_true, y_proba)
    pr_auc = average_precision_score(y_true, y_proba)

    print(f"\n=== RESULTADOS ({nome_modelo}) ===")
    print(f"Accuracy : {acc:.6f}")
    print(f"Precision: {prec:.6f}")
    print(f"Recall   : {rec:.6f}")
    print(f"F1-Score : {f1:.6f}")
    print(f"ROC-AUC  : {roc_auc:.6f}")
    print(f"PR-AUC   : {pr_auc:.6f}")
    print(f"Tempo(s) : {tempo_treino:.2f}")

    print("\nRelatório de Classificação:")
    print(classification_report(y_true, y_pred, digits=6))

    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot()
    plt.title(f"Matriz de Confusão - {nome_modelo}")
    plt.show()

    fpr, tpr, _ = roc_curve(y_true, y_proba)
    roc_auc_plot = auc(fpr, tpr)

    plt.figure(figsize=(6,4))
    plt.plot(fpr, tpr, label=f"{nome_modelo} (AUC = {roc_auc_plot:.4f})")
    plt.plot([0, 1], [0, 1], "--")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"Curva ROC - {nome_modelo}")
    plt.legend(loc="lower right")
    plt.show()

    return {
        "Modelo": nome_modelo,
        "Accuracy": acc,
        "Precision": prec,
        "Recall": rec,
        "F1-Score": f1,
        "ROC-AUC": roc_auc,
        "PR-AUC": pr_auc,
        "Tempo (s)": tempo_treino
    }

# ============================================================
# 15) Ajuste do CCNN para número flexível de atributos
# ============================================================
def vetor_para_matriz_flexivel(X_array, tamanho=6):
    n_amostras = X_array.shape[0]
    total_elementos = tamanho * tamanho
    X_img = np.zeros((n_amostras, tamanho, tamanho), dtype=np.float32)

    for i in range(n_amostras):
        vetor = X_array[i]
        if len(vetor) >= total_elementos:
            vetor_ajustado = vetor[:total_elementos]
        else:
            vetor_ajustado = np.pad(vetor, (0, total_elementos - len(vetor)), mode="constant")
        X_img[i] = vetor_ajustado.reshape(tamanho, tamanho)

    return X_img

X_train_ccnn = vetor_para_matriz_flexivel(X_train_smote, tamanho=6)
X_test_ccnn = vetor_para_matriz_flexivel(X_test_scaled, tamanho=6)

X_train_ccnn = np.expand_dims(X_train_ccnn, axis=-1)
X_test_ccnn = np.expand_dims(X_test_ccnn, axis=-1)

print("\nX_train_ccnn:", X_train_ccnn.shape)
print("X_test_ccnn :", X_test_ccnn.shape)

# ============================================================
# 16) Modelo CCNN
# ============================================================
def construir_ccnn():
    model = models.Sequential([
        layers.Input(shape=(6, 6, 1)),
        layers.Conv2D(32, (2, 2), activation="relu", padding="same"),
        layers.Conv2D(64, (2, 2), activation="relu", padding="same"),
        layers.Dropout(0.2),
        layers.Conv2D(64, (2, 2), activation="relu", padding="same"),
        layers.Flatten(),
        layers.Dense(64, activation="relu"),
        layers.Dropout(0.3),
        layers.Dense(32, activation="relu"),
        layers.Dense(1, activation="sigmoid")
    ])

    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
        loss="binary_crossentropy",
        metrics=[
            "accuracy",
            tf.keras.metrics.Precision(name="precision"),
            tf.keras.metrics.Recall(name="recall"),
            tf.keras.metrics.AUC(name="auc")
        ]
    )
    return model

ccnn_model = construir_ccnn()
ccnn_model.summary()

early_stop = callbacks.EarlyStopping(
    monitor="val_loss",
    patience=5,
    restore_best_weights=True
)

reduce_lr = callbacks.ReduceLROnPlateau(
    monitor="val_loss",
    factor=0.5,
    patience=3,
    verbose=1
)

inicio_treino_ccnn = time.time()

history_ccnn = ccnn_model.fit(
    X_train_ccnn,
    y_train_smote,
    validation_split=0.2,
    epochs=20,
    batch_size=256,
    callbacks=[early_stop, reduce_lr],
    verbose=1
)

tempo_treino_ccnn = time.time() - inicio_treino_ccnn

y_proba_ccnn = ccnn_model.predict(X_test_ccnn).ravel()
y_pred_ccnn = (y_proba_ccnn >= 0.5).astype(int)

resultado_ccnn_dict = calcular_metricas(
    y_test, y_pred_ccnn, y_proba_ccnn, tempo_treino_ccnn, "CCNN"
)

# gráficos do CCNN
plt.figure(figsize=(6,4))
plt.plot(history_ccnn.history["loss"], label="train_loss")
plt.plot(history_ccnn.history["val_loss"], label="val_loss")
plt.title("Loss - CCNN")
plt.legend()
plt.show()

plt.figure(figsize=(6,4))
plt.plot(history_ccnn.history["accuracy"], label="train_acc")
plt.plot(history_ccnn.history["val_accuracy"], label="val_acc")
plt.title("Accuracy - CCNN")
plt.legend()
plt.show()

# ============================================================
# 17) CFTNet ADAPTADO
# ============================================================
def gerar_contrafactual_simples(X_fraude, n_amostras=None, ruido=0.02):
    if n_amostras is None:
        n_amostras = len(X_fraude)

    idx = np.random.choice(len(X_fraude), size=n_amostras, replace=True)
    base = X_fraude[idx].copy()
    perturbacao = np.random.normal(0, ruido, base.shape)
    X_cf = base + perturbacao
    X_cf = np.clip(X_cf, 0, 1)
    return X_cf

def build_cftnet(input_dim):
    entradas = Input(shape=(input_dim,))
    x = Reshape((input_dim, 1))(entradas)

    x_proj = Dense(32)(x)

    attn_output = MultiHeadAttention(num_heads=2, key_dim=16)(x_proj, x_proj)
    x = Add()([x_proj, attn_output])
    x = LayerNormalization()(x)

    ff = Dense(64, activation="relu")(x)
    ff = Dense(32)(ff)
    x = Add()([x, ff])
    x = LayerNormalization()(x)

    x = GlobalAveragePooling1D()(x)
    x = Dense(64, activation="relu")(x)
    x = Dropout(0.3)(x)
    saida = Dense(1, activation="sigmoid")(x)

    model = Model(inputs=entradas, outputs=saida)

    model.compile(
        optimizer="adam",
        loss="binary_crossentropy",
        metrics=[
            "accuracy",
            tf.keras.metrics.Precision(name="precision"),
            tf.keras.metrics.Recall(name="recall"),
            tf.keras.metrics.AUC(name="auc")
        ]
    )
    return model

X_fraude = X_train_smote[y_train_smote == 1]
X_cf = gerar_contrafactual_simples(X_fraude, n_amostras=len(X_fraude), ruido=0.02)
y_cf = np.ones(len(X_cf), dtype=int)

X_train_cft = np.vstack([X_train_smote, X_cf])
y_train_cft = np.hstack([y_train_smote, y_cf])

print("\nX_train_cft:", X_train_cft.shape)
print("y_train_cft:", y_train_cft.shape)

cftnet_model = build_cftnet(X_train_cft.shape[1])
cftnet_model.summary()

early_stop_cft = callbacks.EarlyStopping(
    monitor="val_loss",
    patience=5,
    restore_best_weights=True
)

reduce_lr_cft = callbacks.ReduceLROnPlateau(
    monitor="val_loss",
    factor=0.5,
    patience=3,
    verbose=1
)

inicio_treino_cft = time.time()

history_cft = cftnet_model.fit(
    X_train_cft,
    y_train_cft,
    validation_split=0.2,
    epochs=20,
    batch_size=256,
    callbacks=[early_stop_cft, reduce_lr_cft],
    verbose=1
)

tempo_treino_cft = time.time() - inicio_treino_cft

y_proba_cft = cftnet_model.predict(X_test_scaled).ravel()
y_pred_cft = (y_proba_cft >= 0.5).astype(int)

resultado_cftnet_dict = calcular_metricas(
    y_test, y_pred_cft, y_proba_cft, tempo_treino_cft, "CFTNet"
)

# gráficos do CFTNet
plt.figure(figsize=(6,4))
plt.plot(history_cft.history["loss"], label="train_loss")
plt.plot(history_cft.history["val_loss"], label="val_loss")
plt.title("Loss - CFTNet")
plt.legend()
plt.show()

plt.figure(figsize=(6,4))
plt.plot(history_cft.history["accuracy"], label="train_acc")
plt.plot(history_cft.history["val_accuracy"], label="val_acc")
plt.title("Accuracy - CFTNet")
plt.legend()
plt.show()

# ============================================================
# 18) SICRNN ADAPTADO
# ============================================================
k_features = min(20, X_train_smote.shape[1])

selector = SelectKBest(score_func=f_classif, k=k_features)
X_train_sel = selector.fit_transform(X_train_smote, y_train_smote)
X_test_sel = selector.transform(X_test_scaled)

X_train_sicrnn = X_train_sel.reshape((X_train_sel.shape[0], X_train_sel.shape[1], 1))
X_test_sicrnn = X_test_sel.reshape((X_test_sel.shape[0], X_test_sel.shape[1], 1))

print("\nX_train_sicrnn:", X_train_sicrnn.shape)
print("X_test_sicrnn :", X_test_sicrnn.shape)

def build_sicrnn(input_shape):
    model = Sequential([
        layers.Input(shape=input_shape),
        GRU(64, return_sequences=False),
        Dropout(0.3),
        Dense(32, activation="relu"),
        Dropout(0.2),
        Dense(1, activation="sigmoid")
    ])

    model.compile(
        optimizer="adam",
        loss="binary_crossentropy",
        metrics=[
            "accuracy",
            tf.keras.metrics.Precision(name="precision"),
            tf.keras.metrics.Recall(name="recall"),
            tf.keras.metrics.AUC(name="auc")
        ]
    )
    return model

sicrnn_model = build_sicrnn((X_train_sicrnn.shape[1], X_train_sicrnn.shape[2]))
sicrnn_model.summary()

early_stop_sic = callbacks.EarlyStopping(
    monitor="val_loss",
    patience=5,
    restore_best_weights=True
)

reduce_lr_sic = callbacks.ReduceLROnPlateau(
    monitor="val_loss",
    factor=0.5,
    patience=3,
    verbose=1
)

inicio_treino_sic = time.time()

history_sic = sicrnn_model.fit(
    X_train_sicrnn,
    y_train_smote,
    validation_split=0.2,
    epochs=20,
    batch_size=256,
    callbacks=[early_stop_sic, reduce_lr_sic],
    verbose=1
)

tempo_treino_sic = time.time() - inicio_treino_sic

y_proba_sic = sicrnn_model.predict(X_test_sicrnn).ravel()
y_pred_sic = (y_proba_sic >= 0.5).astype(int)

resultado_sicrnn_dict = calcular_metricas(
    y_test, y_pred_sic, y_proba_sic, tempo_treino_sic, "SICRNN"
)

# gráficos do SICRNN
plt.figure(figsize=(6,4))
plt.plot(history_sic.history["loss"], label="train_loss")
plt.plot(history_sic.history["val_loss"], label="val_loss")
plt.title("Loss - SICRNN")
plt.legend()
plt.show()

plt.figure(figsize=(6,4))
plt.plot(history_sic.history["accuracy"], label="train_acc")
plt.plot(history_sic.history["val_accuracy"], label="val_acc")
plt.title("Accuracy - SICRNN")
plt.legend()
plt.show()

# ============================================================
# 19) TABELA FINAL
# ============================================================
resultados_finais = pd.DataFrame([
    resultado_ccnn_dict,
    resultado_cftnet_dict,
    resultado_sicrnn_dict
])

print("\n=== TABELA FINAL DE COMPARAÇÃO ===")
display(resultados_finais.round(6))

# ============================================================
# 20) EXPORTAR CSV
# ============================================================
resultados_finais.to_csv("comparacao_modelos_sparkov.csv", index=False)
print("\nArquivo salvo: comparacao_modelos_sparkov.csv")
```
